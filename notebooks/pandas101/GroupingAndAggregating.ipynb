```python
import pandas as pd
```

```python
# onto getting mean, median, and mode (utilizing grouping and aggregating! as Denmark single-handedly and confidently demonstrated to me haha)

df = pd.read_csv('../data/TS_spotify.csv')
df
```

```python
mean = df.groupby("album").agg({"popularity" : "mean"})
mean
```

```python
median = df.groupby("album").agg({"popularity" : "median"})
median
```

```python
mode = df.groupby("album")["popularity"].agg(pd.Series.mode)
mode
```

```python
standardDeviation = df.groupby("album").agg({"popularity" : "var"})
standardDeviation
```
